```python
import requests
from bs4 import BeautifulSoup

#Initialisation des variables a récupérer 
ls_titre = []
ls_image = []
ls_note = []
ls_prix = []
ls_description = [] 
ls_product_description = []
ls_upc = []
ls_product_type = []
ls_taxe = []
ls_availability = []
ls_number_of_reviews = []

ls_des_liens_url_des_livres = []
page = 1

while True:

  url = 'https://books.toscrape.com/catalogue/page-'+ str(page)+'.html'
  
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  
  blocContentantUrlDuLivre = soup.find_all('div', class_="image_container")
  if blocContentantUrlDuLivre == []:
    print('fin')
    break
  else:
    for diva in blocContentantUrlDuLivre:
      #print("https://books.toscrape.com/catalogue/"+  str(diva.find('a')['href']))
      ls_des_liens_url_des_livres.append("https://books.toscrape.com/catalogue/"+str(diva.find('a')['href']))
      
    page += 1

#Boucler sur les 1000 liens des pages de chaque livre pour récuperer le detail de chaque livre
for lien_dun_livre in ls_des_liens_url_des_livres:
  responseDuLivre = requests.get(lien_dun_livre)
  soupDuLivre = BeautifulSoup(responseDuLivre.text, 'html.parser')
  #Recuperation des titres
  parentDuTitre = soupDuLivre.find('div', class_='col-sm-6 product_main')
  ls_titre.append(parentDuTitre.find("h1" , recursive=False).text)
  #Recuperation des images
  parentDeImage = soupDuLivre.find('div', class_='item active')
  ls_image.append(parentDeImage.find("img" , recursive=False).get('src'))
  #Recuperation des notes
  parentDeNote = soupDuLivre.find('div', class_='col-sm-6 product_main')
  filsDeParentNote = parentDeNote.findChild('p',class_='star-rating')
  valeurDeNote = filsDeParentNote.attrs.get('class')[1]
  ls_note.append(valeurDeNote)
  #Recuperation des prix
  parentDuPrix = soupDuLivre.find('p', class_='price_color')
  ls_prix.append(parentDuPrix.text)
  #Recuperation de la description
  parentDeLaDescription = soupDuLivre.find('article', class_='product_page')
  ls_description.append(parentDeLaDescription.findChild('p').text)
  #Recuperation de upc
  enfantsDeLaTable = soupDuLivre.find('table', class_='table table-striped').findChildren('td') 
  ls_upc.append(enfantsDeLaTable[0].text)
  #Recuperation de product_type
  ls_product_type.append(enfantsDeLaTable[1].text)
  #Recuperation de la taxe
  ls_taxe.append(enfantsDeLaTable[4].text)
  #Recuperation de availability
  ls_availability.append(enfantsDeLaTable[5].text)
  #Recuperation du nombre de vues
  ls_number_of_reviews.append(enfantsDeLaTable[6].text)
```

```python
import pandas as pd
pd.reset_option('max_columns')
df = pd.DataFrame({'TITRE':ls_titre, 'IMAGE':ls_image, 'NOTES':ls_note, 'PRIX':ls_prix, 'DESCRIPTION':ls_description, 'UPC':ls_upc, 'TYPE_DE_PRODUIT':ls_product_type, 'TAXE':ls_taxe, 'DISPONIBILITES':ls_availability, 'NOMBRE_DE_VUES':ls_number_of_reviews })
```
