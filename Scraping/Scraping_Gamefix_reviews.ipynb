## Partiel Scraping 
Récupérer sur [trustpilot.com](https://www.trustpilot.com/review/gamefix.fr) les informations suivantes sur les commentaires des clients Gamefix :
- Nombres d’avis du client
- La localisation du client
- La note
- L’avis


## Imports


```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
```

## Recuperation des données pour tous les commentaires d'une page : fonctionnel

```python
url = "https://www.trustpilot.com/review/gamefix.fr"

response = requests.get(url)
response_html = BeautifulSoup(response.text, 'html.parser')

review = response_html.find_all('div', class_="paper_paper__1PY90 paper_square__lJX8a card_card__lQWDv card_noPadding__D8PcU styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ") 

for item in review:
  client_name = item.find('div', class_= "typography_typography__QgicV typography_bodysmall__irytL typography_weight-medium__UNMDK typography_fontstyle-normal__kHyN3 styles_consumerName__dP8Um").text # stock le nom du client
  title = item.find('h2', class_="typography_typography__QgicV typography_h4__E971J typography_color-black__5LYEn typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3 styles_reviewTitle__04VGJ").text #stock le titre du commentaire
  paragraph = item.find('p', class_="typography_typography__QgicV typography_body__9UBeQ typography_color-black__5LYEn typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3") #stocke le paragraphe du commentaire
  paragraph = str(paragraph)
  score = item.find('div', class_="star-rating_starRating__4rrcf star-rating_medium__iN6Ty") #stock la note du commentaire
  score = str(score)
  text_to_delete = {'<div class="star-rating_starRating__4rrcf star-rating_medium__iN6Ty"><img alt="':'',                     #je cree une liste key:value des mots que je veux remplacer(key) par le vide (value) de mon str pour les var paragraph et score
                  '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-5.svg"/></div>':'',
                  '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-4.svg"/></div>':'',
                  '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-3.svg"/></div>':'',
                  '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-2.svg"/></div>':'',
                  '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-1.svg"/></div>':'',
                  '<br/>':'','</p>':'',
                  '<p class="typography_typography__QgicV typography_body__9UBeQ typography_color-black__5LYEn typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3" data-service-review-text-typography="true">':''}
  for key, value in text_to_delete.items():                                       #je parcours ma liste
    score = score.replace(key, value)             #je remplace les mots a supprimer(key) par le vide (value) dans mon string
    paragraph = paragraph.replace(key, value)

  location = item.find('div', class_="typography_typography__QgicV typography_bodysmall__irytL typography_color-gray-6__TogX2 typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3 styles_detailsIcon__Fo_ua").text #stock la localisation du client
  client_review_count = item.find('span', class_="typography_typography__QgicV typography_bodysmall__irytL typography_color-gray-6__TogX2 typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3").text #stock le nombre de review fait par un client
  print(location)

 
```

## Recuperation des infos pour tous les commentaires de toutes les pages : fonctionnel

```python
page = 1
#Initialisation des variables a récupérer 
ls_client_name = [] #stocke le nom des client
ls_title = [] #stocke le titre du commentaire
ls_paragraph = [] #stocke le texte du commentaire
ls_score = [] #stocke le score 
ls_location = [] #stocke la localisation
ls_client_review_count = [] #stocke le nombre de commentaire fait par un client

while True:
  print(page)

  url ='https://www.trustpilot.com/review/gamefix.fr?languages=all&page='+ str(page)

  response = requests.get(url)

  if response.status_code != 200:
    break

  soup = BeautifulSoup(response.text, 'html.parser')

  review = soup.find_all('div', class_="paper_paper__1PY90 paper_square__lJX8a card_card__lQWDv card_noPadding__D8PcU styles_cardWrapper__LcCPA styles_show__HUXRb styles_reviewCard__9HxJJ") 

  for item in review:
    client_name = item.find('div', class_= "typography_typography__QgicV typography_bodysmall__irytL typography_weight-medium__UNMDK typography_fontstyle-normal__kHyN3 styles_consumerName__dP8Um").text # stock le nom du client
    ls_client_name.append(client_name)
    
    title = item.find('h2', class_="typography_typography__QgicV typography_h4__E971J typography_color-black__5LYEn typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3 styles_reviewTitle__04VGJ").text #stock le titre du commentaire
    ls_title.append(title)
    
    paragraph = item.find('p', class_="typography_typography__QgicV typography_body__9UBeQ typography_color-black__5LYEn typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3") #stocke le paragraphe du commentaire
    paragraph = str(paragraph)
    score = item.find('div', class_="star-rating_starRating__4rrcf star-rating_medium__iN6Ty") #stock la note du commentaire
    score = str(score)
    text_to_delete = {'<div class="star-rating_starRating__4rrcf star-rating_medium__iN6Ty"><img alt="':'',                     #je cree une liste key:value des mots que je veux remplacer(key) par le vide (value) de mon str pour les var paragraph et score
                    '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-5.svg"/></div>':'',
                    '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-4.svg"/></div>':'',
                    '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-3.svg"/></div>':'',
                    '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-2.svg"/></div>':'',
                    '" src="https://cdn.trustpilot.net/brand-assets/4.1.0/stars/stars-1.svg"/></div>':'',
                    '<br/>':'','</p>':'',
                    '<p class="typography_typography__QgicV typography_body__9UBeQ typography_color-black__5LYEn typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3" data-service-review-text-typography="true">':''}
    for key, value in text_to_delete.items():                                       #je parcours ma liste
      score = score.replace(key, value)             #je remplace les mots a supprimer(key) par le vide (value) dans mon string
      paragraph = paragraph.replace(key, value)

    ls_paragraph.append(paragraph)
    ls_score.append(score)
    
    location = item.find('div', class_="typography_typography__QgicV typography_bodysmall__irytL typography_color-gray-6__TogX2 typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3 styles_detailsIcon__Fo_ua").text #stock la localisation du client
    ls_location.append(location)
    
    client_review_count = item.find('span', class_="typography_typography__QgicV typography_bodysmall__irytL typography_color-gray-6__TogX2 typography_weight-regular__TWEnf typography_fontstyle-normal__kHyN3").text #stock le nombre de review fait par un client
    ls_client_review_count.append(client_review_count)
  
  page += 1
```

## To csv


```python
pd.reset_option('max_columns')
df = pd.DataFrame({'CLIENT NAME':ls_client_name,'CLIENT_LOCATION': ls_location, 'REVIEW_TITLE':ls_title, 'REVIEW_PARAGRAPH':ls_paragraph, 'REVIEW_SCORE':ls_score, 
                   'CLIENT_REVIEW_COUNT':ls_client_review_count})
df
```

```python
df.to_csv(r'\Gamefix_Reviews_Infos.csv')
```
