#### Data Engineering

* Suppression de _c0
* Caster les élements carat, depth, table, price, x, y et z  **en entier** ou **en float**

```
%sql

drop table if exists d2; -- Suppression de la table d2

create table d2 as -- Creation de la table cleaned
select float(carat) as carat -- Casting de la data str en int et renommage de la variable
, cut
, color
, clarity
, float(depth) as depth
, int(table) as table
, int(price) as price
, float(x) as x
, float(y) as y
, float(z) as z
from diamond
```

#####  2. Calculer le prix minimum, le prix maximum, le prix moyen en arrondissant à 2 après la virgule pour l'ensemble des diamants

```
%sql
select min(price) as min_price  -- Prix minimum
, max(price) as max_price -- Prix maximum des diamants
, avg(price) as moy_price  -- Prix moyen des diamnts
, round(avg(price), 2) as moy_price_round -- Prix moyen arrondi
from d2
```

```
%sql
select * from d2 order by price desc
```

##### Calculer le prix minimum, le prix maximum, le prix moyen en arrondissant à 2 après la virgule **par couleur**

* Par: group by
* Splitter par **color**
* Aggregation: **max, min, avg**
* Variable aggrégéz: **Price**

```
%sql
select color
, min(price) as min_price
, max(price) as max_price
, round(avg(price), 2) as moy_price_r
from d2
group by color -- Group par couleur
order by color -- Trie par ordre alphabétique
```
